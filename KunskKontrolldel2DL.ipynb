```python
#Importing libraries
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model
from time import sleep
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the model and xml-file as a Cascade classifier for face detection
face_classifier = cv2.CascadeClassifier(r'C:\Users\niksan2\Downloads\haarcascade_frontalface_default.xml')
classifier = load_model(r'C:\Users\niksan2\Downloads\model_best.h5')

#Setting the emotion detection labels in alphabetical order
emotion_labels = ['Angry','Disgust','Fear','Happy','Neutral', 'Sad', 'Surprise']
```

```python
#Compiling the model
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Accessing the webcamera
cap = cv2.VideoCapture(0)
```

```python
ret, frame = cap.read()
```

```python
#

while True:
    ret, frame = cap.read()
    labels = []
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray) #converting webcamera image to gray scale as model was trained on gray scale images 

    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2) #drawing a yellow rectangle around each detected face 
        roi_gray = gray[y:y+h,x:x+w] 
        roi_gray = cv2.resize(roi_gray,(48,48),interpolation=cv2.INTER_AREA) #Extracting the region of interest of the gray scale image and resizing it 
                                                                            #to 48x48 pixels, the size of the images during training



        if np.sum([roi_gray])!=0: #Ensuring the region of interest is not empty by checking the sum of the pixels not being equal to zero
            roi = roi_gray.astype('float')/255.0
            roi = img_to_array(roi)
            roi = np.expand_dims(roi,axis=0) #converting the region of interest to an array with the same dimensions as the needed input shape for the model

            prediction = classifier.predict(roi)[0] #Using the loaded model to predict emotions on the face in real time,
            label=emotion_labels[prediction.argmax()] #the corresponding class to the highest probability will result in the emotion label
            label_position = (x,y) #Setting the position of the label on the frame
            cv2.putText(frame,label,label_position,cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2) #Using green color for the label text
        else:
            cv2.putText(frame,'No Faces',(30,80),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2) #If region of interest is empty, No Faces will be displayed in green color
    cv2.imshow('Emotion Recognition',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): #Using the key 'q' to quit and break the loop 
        break

cap.release()
cv2.destroyAllWindows()
```
